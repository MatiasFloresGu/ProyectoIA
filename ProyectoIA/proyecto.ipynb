```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Laberinto (1 = pared, 0 = camino)
maze_array = np.array(maze)

# Posición inicial del ratón
mouse_position = [1, 1]  # Fila, columna inicial (ajusta según tu laberinto)

# Movimientos simulados (puedes reemplazar con un algoritmo real)
movements = [
    [1, 2], [1, 3], [2, 3], [3, 3], [3, 4], [3, 5],
    [4, 5], [5, 5], [5, 6], [5, 7], [4, 7], [3, 7]
]

# Configuración inicial de la gráfica
plt.figure(figsize=(10, 10))
plt.title("Recorrido del Ratón en el Laberinto")

for move in movements:
    # Actualizar la posición del ratón
    maze_array[mouse_position[0], mouse_position[1]] = 0  # Dejar el camino libre
    mouse_position = move
    maze_array[mouse_position[0], mouse_position[1]] = 2  # Marcar la posición del ratón
    
    # Mostrar el laberinto
    plt.imshow(maze_array, cmap="cool")  # Cambia el mapa de colores si prefieres otro
    plt.xticks([])  # Ocultar marcas de los ejes x
    plt.yticks([])  # Ocultar marcas de los ejes y
    plt.grid(visible=False)  # Ocultar la cuadrícula
    plt.pause(0.5)  # Pausa para simular el movimiento (en segundos)

# Detener modo interactivo y mostrar el resultado final
plt.ioff()
plt.show()
```
